**Exploratory Data Analysis- Terrorism(1970-2017)**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('/content/gdrive/MyDrive/Global Terrorism - START data/globalterrorismdb_0718dist.csv',encoding='Windows-1252')
```

```python
dataset.shape
```

```python
column_name = dataset.columns
```

```python
column_name.tolist()
```

```python
dataset.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'state','region_txt':'Region','attacktype1_txt':'AttackType',
                     'target1':'Target','nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                     'weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
```

```python
new_dataset=dataset[['Year','Month','Day','Country','state','Region','city','latitude','longitude','AttackType','Killed','Wounded','Target','Summary','Group','Target_type','Weapon_type','Motive']]
```

```python
new_dataset.shape
```

```python
new_dataset.hist(figsize=(20,20))
plt.show()
```

```python
new_dataset.describe()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(new_dataset.corr(), annot=True,fmt='.0%', cmap = 'viridis')
plt.show()
```

```python
plt.figure(figsize=(20,10))

sns.countplot(y='Year', data=new_dataset)
plt.title('Terrorist Activities by each Year')

plt.show()
```

In the year, 2014 and 2015 large number of Terrorist activities observed.

```python
pd.crosstab(new_dataset.Year, new_dataset.Region).plot(kind='area',figsize=(20,12))
plt.title('Terrorist Activities by Region in each Year')
plt.ylabel('Number of Attacks')
plt.show()
```

In Middle east and North Africa region the terrorist activities Increases.

```python
import numpy as np
Year=new_dataset.Year.value_counts().to_dict()
rate=((Year[2017]-Year[1970])/Year[2017])*100
print(Year[1970],'attacks happened in 1970 &',Year[2017],'attacks happened in 2017')
print('So the number of attacks from 1970 has increased by',np.round(rate,0),'% till 2017')
```

```python
plt.figure(figsize=(15,10))
sns.countplot(x=pd.Series(new_dataset['AttackType']), order=new_dataset['AttackType'].value_counts().index, palette='hls')

plt.xticks(rotation=90)
plt.xlabel('Method')
plt.title('Method of Attack')
plt.show()
```

**The highest record for Method of attacks observed is Bombing/Explosion.**

```python

plt.figure(figsize=(15,10))
sns.countplot(x=pd.Series(new_dataset['Target_type']), order=new_dataset['Target_type'].value_counts().index, palette='hot')

plt.xticks(rotation=90)
plt.xlabel('Target Type')
plt.title('Target Type')
plt.show()
```

**The target for attacks were in Private Citizens and property .**

```python
fig,axes = plt.subplots(figsize=(20,15),nrows=1,ncols=2)
sns.barplot(x = new_dataset['Country'].value_counts()[:20].values, y = new_dataset['Country'].value_counts()[:20].index,
            ax=axes[0],palette = 'magma');
axes[0].set_title('Terrorist Attacks per Country')
sns.barplot(x= new_dataset['Region'].value_counts().values,y=new_dataset['Region'].value_counts().index,
            ax=axes[1])
axes[1].set_title('Terrorist Attacks per Region')
fig.tight_layout()
plt.show()
```

```python
number_of_occurences=new_dataset.groupby(['Country'],as_index=False).count()

```

```python
max_count=number_of_occurences['Year'].max()
max_id=number_of_occurences['Year'].idxmax() #idmax () returns the index of the (first occurrence of the) maximum value
max_name=number_of_occurences['Country'][max_id]
print(max_name, 'has suffered the maximum number of terrors effect of', max_count)
```

```python
min_count=number_of_occurences['Year'].min()
min_id=number_of_occurences['Year'].idxmin()
min_name=number_of_occurences['Country'][min_id]
print(min_name, 'has suffered the minimum number of terrors effect of',min_count)
```

```python
plt.subplots(figsize=(11,15))
sns.barplot(y=new_dataset['Group'].value_counts()[1:12].index,x=new_dataset['Group'].value_counts()[1:12].values,
           palette='flare')
plt.title('Most Active Terrorist Organizations')
plt.show()

```

Most Active Terrorist Organizations were Taliban and Islamic state of Iraq and the levant(ISIL).
